### Basic Sentiment : 😀 /  😕 / 😠 / 😟

Bunu göz önünde bulundurarak, bir model belirtmeden duygu analizi için bir pipeline oluşturun ve "Transformers are awesome!" cümlesiyle bu modeli test etmeye çalışın. Diğer cümleleri de denemekten çekinmeyin.

```python
from transformers import pipeline

# Herhangi bir model belirtmezsek varsayılan olarak 'distilbert-base-uncased-finetuned-sst-2-english' indirilir
sentiment_pipeline = pipeline("sentiment-analysis")

result = sentiment_pipeline("Transformers are awesome!")
print(f"Standart Model Sonucu: {result}")
```

### Nüanslı Duygu 🤔

HuggingFace, bir model belirtmezsek varsayılan olarak `distilbert-base-uncased-finetuned-sst-2-english` modelini kullanır. Bu model birçok temel kullanım durumunda iyi sonuç verir, ancak oldukça sınırlı bir metin külliyatı üzerinde eğitildiği için: 

`Stanford Sentiment Treebank, dildeki duyguların kompozisyonel etkilerinin tam bir analizini mümkün kılan, tamamen etiketlenmiş ayrıştırma ağaçlarına sahip bir metin kümesidir. Bu metin kümesi, Pang ve Lee (2005) tarafından tanıtılan veri setine dayanmaktadır ve film eleştirilerinden çıkarılan 11.855 tek cümleden oluşmaktadır. Stanford ayrıştırıcı ile ayrıştırılmış ve bu ayrıştırma ağaçlarından toplam 215.154 benzersiz ifade içermektedir. Her ifade, 3 insan yargıç tarafından açıklanmıştır.`

Bu modelin, modern dil içeren cümlelerde muhtemelen kötü performans göstereceği oldukça açıktır: örneğin, "These beats are sick!" . Bu cümleleri şimdi boru hattınızda çalıştırmayı deneyin, oldukça olumlu duygular ifade etmelerine rağmen negatif puanlar alacaksınız.

HuggingFace modellerinin listesine gidin ve Twitter duygularını uzmanlaşmış bir model bulup bulamayacağınızı kontrol edin (`“twitter-roberta-base-sentiment-latest”` aramak iyi bir başlangıç olabilir) - umarım bu yeni terimler ile biraz daha güncel olur! Şimdi ikinci bir boru hattı oluşturun, bu sefer kullanmak istediğimiz modeli __belirtin__ (`model=` kullanın) ve ince ayarlı bir model kullandığımız için performansımızın anında nasıl iyileştiğini görün.

```python
# Standart model "sick" kelimesini görünce cümleyi "Negatif" sanacaktır:
bad_result = sentiment_pipeline("These beats are sick!")
print(f"Standart Modelin 'sick' (harika) kelimesi yanılgısı: {bad_result}")

# Şimdi Twitter argosunu ve yeni dili anlayan 'twitter-roberta-base-sentiment-latest' modelini kullanıyoruz
twitter_sentiment_pipeline = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment-latest")

good_result = twitter_sentiment_pipeline("These beats are sick!")
print(f"\nTwitter Roberta Modelinin Sonucu (Artık doğru anlıyor): {good_result}")
```

### Diğer dillerde duygu

İlk pipeline'ımız diğer dillerdeki basit cümlelerde (örneğin “C'est bon” veya “Esta bueno”) şaşırtıcı derecede iyi performans gösterse de, bu dillerdeki daha karmaşık fikirleri işlerken başarısız oluyor. 

İşte Jurassic World Dominion filmi için bir örnek yorum 😬:

"This was frankly a spectacular failure from start to finish, with  remarkably uninspired performances from some very well-paid actors who acted with all the passion of a wet biscuit" 

Koreceye çevrildiğinde şöyle okunur: "이것은 솔직히 처음부터 끝까지 엄청난 실패였으며 젖은 비스킷의 모든 열정으로 연기한 일부 매우 보수가 좋은 배우들의 현저하게 영감을 받지 못한 연기로 끝났습니다."

Korece metni Twitter modelinizden geçirin; yorumun ne kadar kötü olduğunu fark etmeyeceklerini göreceksiniz. 

Şimdi HuggingFace kütüphanesinde daha iyi performans gösterebilecek bir model bulup kullanmayı deneyin. `“matthewburke/korean_sentiment”` modelini bir `text-classification` boru hattında kullanmayı deneyin ve sonuçlarınızın değişip değişmediğini kontrol edin.

```python
korean_text = "이것은 솔직히 처음부터 끝까지 엄청난 실패였으며 젖은 비스킷의 모든 열정으로 연기한 일부 매우 보수가 좋은 배우들의 현저하게 영감을 받지 못한 연기로 끝났습니다."

# Twitter modeli bu dili bilmediği için rastgele bir cevap verir (Muhtemelen Neutral)
twitter_korean_result = twitter_sentiment_pipeline(korean_text)
print(f"İngilizce Modelin Korece Metne Tepkisi: {twitter_korean_result}")

# Özellikle Korece için eğitilmiş modeli kullanıyoruz
korean_pipeline = pipeline("text-classification", model="matthewburke/korean_sentiment")

real_korean_result = korean_pipeline(korean_text)
print(f"\nGerçek Korece Duygu Modeli Sonucu (0=Negatif, 1=Pozitif): {real_korean_result}")
```

### Özetleme

Bir başka gerçekten yararlı NLP görevi, büyük miktarda bilgiyi çok az sayıda kelimeyle özetlemektir. BART, özetleme gibi görevlerde iyi performans gösteren bir modeldir; derslerde kısaca gördüğünüz iki modelin birleşiminden oluşur: BERT modeli ve otogresif stil GPT modeli. Bu model hakkında daha fazla bilgi için [bu bağlantıya](https://www.projectpro.io/article/transformers-bart-model-explained/553) göz atın. 

BART modelleri oldukça büyük olabileceğinden, HuggingFace'de mevcut en küçük damıtma modellerinden biri olan `distilbart-xsum-12-6` modelini bulmaya çalışın (damıtma hakkında daha sonra daha fazla konuşacağız!). Bu modeli bir `“özetleme”` boru hattına entegre edin, ardından bir metin alın (örneğin, [bir BBC makalesini](https://www.bbc.com/news/topics/cx2pk70323et) kopyalayıp yapıştırarak) ve boru hattınızla özetleyin!

Not: Bağlam pencerelerine dikkat etmelisiniz - burada, girişinizin model için çok uzun olması nedeniyle bir sorunla karşılaşabilirsiniz!

```python
# Herhangi bir makaleden kopyalanmış örnek uzun metin
text = """
Artificial intelligence (AI) is intelligence demonstrated by machines, as opposed to the natural intelligence displayed by animals including humans. AI research has been defined as the field of study of intelligent agents, which refers to any system that perceives its environment and takes actions that maximize its chance of achieving its goals. The term "artificial intelligence" had previously been used to describe machines that mimic and display "human" cognitive skills that are associated with the human mind, such as "learning" and "problem-solving". This definition has since been rejected by major AI researchers who now describe AI in terms of rationality and acting rationally, which does not limit how intelligence can be articulated.
"""

# Özetleme pipeline'ı oluştur (sshleifer damıtılmış küçük model)
summarization_pipeline = pipeline("summarization", model="sshleifer/distilbart-xsum-12-6")

# Metni özetle (max ve min uzunlukları belirtebiliriz)
summary = summarization_pipeline(text, max_length=50, min_length=10, do_sample=False)
print(f"Metin Özeti: {summary[0]['summary_text']}")
```

Nasıl yapılacağını biliyorsunuz: `roberta-base-squad2` modeli ile bir `“soru-cevap”` pipeline oluşturun, ardından daha önce seçtiğiniz `makaleyi` bağlam olarak ekleyin ve bununla ilgili bir `soru` sorun.

```python
# Soru Cevap pipeline'ı
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

# Aynı metin (context) üzerinden soru soruyoruz
question = "Who rejected the previous definition of AI?"

qa_result = qa_pipeline(question=question, context=text)
print(f"Soru: {question}")
print(f"Cevap: {qa_result['answer']} (Güven Skoru: {qa_result['score']:.4f})")
```

### Bonus: Çok modlu olalım 😎: Görsel Soru Yanıtlama

İstersek, görüntüler üzerinde soru-cevap tarzı modeller bile kullanabiliriz. Bu modellerin çoğu, görüntülerden metinleri ayıklayan ve bunları bir dil modeline aktaran zincirleri kullanır. Aşağıdaki modeli kullanmak için, görüntülerimizden metinleri ayıklamamıza yardımcı olacak iki kütüphane olan `pip install Pillow pytesseract` komutunu çalıştırmanız gerekir. 

Bu işlem tamamlandıktan sonra, bir `“document-question-answering”` pipeline oluşturacağız. Bunun için bir modele ihtiyacımız olacak, bu nedenle HuggingFace'de `layoutlm-invoices` modelini arayın. Ardından bununla ilgili sorular sormaya çalışın.   [`receipt.webp`](https://drive.google.com/file/d/1slICDEjI1AXuWXybfSBIt5cCzteFgb3C/view?usp=sharing) (görüntüyü veri klasörünüze indirin veya modeli çağırdığınızda URL'yi doğrudan modele aktarın). Yumurtaların fiyatını, satış vergisinin ne kadar olduğunu ve toplam tutarın ne kadar olduğunu sorun. Kendi görüntülerinizle denemekten çekinmeyin!

Bunun çalışması için bazı bağımlılıklara ihtiyacınız olacak:

```python
# Mac için, yorum satırını kaldırın:
# !brew install tesseract

# Linux veya Colab vb. için, şu satırların yorum satırını kaldırın:
# !sudo apt install tesseract-ocr
# !sudo apt install libtesseract-dev

# Ardından çekirdeğinizi yeniden başlatın ve deneyin!
```

```python
import requests
from PIL import Image
from transformers import pipeline

# 1. Pipeline'ı oluştur
document_qa_pipeline = pipeline("document-question-answering", model="impira/layoutlm-document-qa")

# 2. Hugging Face'in kendi sunucusundaki resmi test faturası
url = "https://huggingface.co/spaces/impira/docquery/resolve/2359223c1837a7587402bda0f2643382a6eefeab/invoice.png"

# 3. Resmi güvenli bir şekilde internetten indirip PIL objesine çeviriyoruz
# (Bu sayede model "bu resim mi değil mi" derdinden kurtuluyor)
resim = Image.open(requests.get(url, stream=True).raw).convert("RGB")

# 4. Sorumuzu soruyoruz
question = "What is the invoice number?"

# 5. URL yerine doğrudan indirdiğimiz resmi modele veriyoruz!
vqa_result = document_qa_pipeline(image=resim, question=question)

print(f"Faturaya Sorulan Soru: {question}")
print(f"Faturadan Bulunan Cevap: {vqa_result[0]['answer']}")
```
